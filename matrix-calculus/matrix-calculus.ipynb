# Matrix Calculus

```python
import numpy as np
import matplotlib.pyplot as plt
import sympy as sp
```

```python
# graphical representation of a gradient 


# function to plot
def f(x, y):
    return  x**2 + y**2

# compute the gradient
def grad_f(x, y):
    return np.array([2 * x, 2*y])

# create a grid of points
x = np.linspace(-2, 2, 20)
y = np.linspace(-2, 2, 20)
X, Y = np.meshgrid(x, y)

# compute the gradient at each point
Gx, Gy = grad_f(X, Y)

# plotting  the gradient field
fig, ax = plt.subplots(figsize=(8, 8))
contours = ax.contour(X, Y, f(X, Y), levels=10, cmap='gray')
ax.quiver(X, Y, Gx, Gy, color='red', pivot='middle', scale=10)

ax.set_title("Gradient Field of f(x, y) = x + y")
ax.set_xlabel("x")
ax.set_ylabel("y")
ax.grid(True)
plt.axis('equal')
plt.show()
```

```python
# calculating a  jacobian 
x, y = sp.symbols('x y')

# define vector valued function
f = sp.Matrix([
    3*x,
    -x*y**2,
    x + y
])

# define the variables vector
vars = sp.Matrix([x, y])

# compute jacobian
J= f.jacobian(vars)

sp.pprint(J)
```
